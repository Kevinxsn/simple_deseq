## Define file path
Here, lets define the file path of all of the data we need, please note that the order need to be corresponded to the column name we defined before

```python
file_list = ['../testfiles/benchmark_data/Chow_Rep1.genes.results', 
             '../testfiles/benchmark_data/Chow_Rep2.genes.results', 
             '../testfiles/benchmark_data/Chow_Rep3.genes.results', 
             '../testfiles/benchmark_data/HFD_Rep1.genes.results', 
             '../testfiles/benchmark_data/HFD_Rep2.genes.results', 
             '../testfiles/benchmark_data/HFD_Rep3.genes.results']
```

## Create a texi_import
Then we can call simle_texi to help us import the data, here is the argument explanation:
`simeple_tximport` takes six arguments:

1. `list_of_def_path`: A Python list containing the paths to the table files.
2. `column_names`: A Python list specifying the column names to use for each table after conversion. It is recommended to use easily identifiable group names, such as `['old1', 'old2', 'young1', 'young2']`.
3. `gene_id_index`: An integer indicating the column number that identifies the gene. For example, if the gene is stored in the second column of the table, this number should be `1` (since the index starts from 0).
4. `count_column_index`: An integer indicating the column number that identifies the counts of each gene.
5. `abundance_column_index`: An integer or `None`, indicating the column number that identifies the abundance of each gene.
6. `length_column_index`: An integer or `None`, indicating the column number that identifies the length of each gene.

```python
txi = simple_teximport.simple_texi(file_list, column_name, gene_id_index=0, count_column_index=4, abundance_column_index=5, length_column_index=3)
print(txi)
```

# Part 2: Filter the data

Now, we need to filter out the data. In the lab, we were ask to filter the data 'to exclude genes with low counts, specifically those not exceeding a threshold of 10 counts in at least two samples, to minimize noise and enhance statistical power'. And we have to do them manually in R, however, in simple_texi, we implement a filter function that can help me do that in one simple call.

```python
txi.filter(threshold=10,min_samples=2)
```

```python
txi.counts
```
